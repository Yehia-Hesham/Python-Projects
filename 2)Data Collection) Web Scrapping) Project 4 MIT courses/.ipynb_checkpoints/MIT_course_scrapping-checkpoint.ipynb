```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrape_courses():
    response = requests.get('https://ocw.mit.edu/courses/electrical-engineering-and-computer-science/')
    soup = BeautifulSoup(response.text, 'html.parser')
    courses = soup.find_all('ul',attrs={'class':'courseRow'})

    with open('courses_scrapped.csv','w',encoding='utf-8')as csv_file:
        fieldnames = ['code','link','title','level']
        writer = csv.DictWriter(csv_file, fieldnames)
        writer.writeheader()
        
        all_courses = []
        for course in courses[1:]:
            course_code = str(course.find('li',attrs={'class':'courseNumCol'}).get_text().strip())
            course_link = 'https://ocw.mit.edu/' + course.find('li',attrs={'class':'courseNumCol'}).find('a').get('href')
            course_title = course.find('li',attrs= {'courseTitleCol'}).get_text().strip()
            course_level = course.find('li',attrs= {'courseLevelCol'}).get_text().strip()
            
            new_course = {'code': course_code, 'link': course_link, 'title': course_title, 'level':course_level}

            all_courses.append(new_course)
            writer.writerow(new_course)

        return all_courses
        print(all_courses)
    #     print(course_code)
    #     print(course_link)
    #     print(course_title)
    #     print(course_level)
```

```python
print(scrape_courses())
```
