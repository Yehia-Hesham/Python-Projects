```python
df['Name'] = df['Name'].apply(lambda x: x.replace(" ",""))
```

```python
df
```

```python
df['texts'] = (df['Name'] + " ")* df['Songs']
```

```python
df['texts']
```

```python
text = ' '.join(df['texts'])
```

```python
text[1:100] #example
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
```

```python
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
